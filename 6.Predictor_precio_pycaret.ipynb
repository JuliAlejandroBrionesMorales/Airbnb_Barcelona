```python
df_pread_new = df[['neighbourhood_cleansed','room_type','minimum_nights','bedrooms','beds','bathrooms', 'price']]
```

```python
setup = setup(df_pread_new, target = 'price') # configuración inicial
```

```python
best = compare_models() # comparamos modelos para poder encontrar el mejor
```

```python
xgboost = create_model('xgboost') # creamos el modelo con las mejores estadísticas
```

```python
tuned_xgboost = tune_model(xgboost) # intentamos ajustar los hiperparámetros del modelos para mejorar el rendimiento
```

```python
evaluate_model (xgboost) # analizamos el rendimiento del modelo medidante varias gráficos y métricas
```
